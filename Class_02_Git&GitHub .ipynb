```python
# Os Commands:-
pwd -> Present working Directory
ls -> lists all files and folders in current directory 
ls -al -> lists all files and folders in current directory in hidden files
cd .. -> Move one directory back 
cd Directory_Name/Path -> Move to specified directory available in CWD
mkdir Directory_Name -> Creates a diecrtory with specified name
mv 'src directory name' 'dst directory name' - moves
rmdir Dir_name - Removes directory


# Git Commands:-
For set up
git config -- global user.name User_name_in_quotation_if_space_in_name
git config -- global user. email User_email_id
git config -- global core.autocrlf true(win)/input(Mac)
git config -- global --list


git clone Repo_link -> Clones the respository from cloud to your local system
git status - to check the status of the repo
git add Dir_name/Filename_with_Extension - Move the file to staging area
git commit - m 'msg' - move the file to local repo with the msg
git push - push the code to cloud repo (Will ask for the credentials while doing first time)
git pull - pulling all the changes from your cloud repo to local repo


# File related operation:-
To open a file from CLI - vim filename.ext (# Only to add a file in CLI - touch filename.py)
to go into insert mode - press 'i'
to save and exit - press - 'esc', :wq
```

```python
1. Create a file into a local repositery - Working Directory 
2. Move the changes to the staging area - add
3. Move the changes to Local Repository (.git) - Commit
4. Then move to Github - Cloud Repository - Push 
```

Different Types of control system:

```python
# Two main types:-
#     Centralized 
#     Decentralized/Distributed

# Centralized:- 
#     Free:- Subversion, CVS
#     Commercial:- Clear case, Perforce, Team Foundation Server (TFS)
#     Required connection for central server for most operations

# Distributed:-
#     Mercurial
#     Git
#     Most operations are local 
#     Central Server not required 
```

Distriubuted Source Control System:-

```python
# Not required to be centralized
# Massively scales
# Open Source 
# Developed for linux project requirments 
# Most operations are local 
# Very fast
# Free and Open Source 
# Active community 
# Most popular DVS, VCS - Git
```

```python
# Git Bash Opersational commands in sequence with details:-

# pwd - It will show the local Desktop location by default (It is present working Directory)
# ls - We will get all the files and folders available by default 
# cd Desktop - By this we will reach at Desktop (cd is used to reach the folder or location required only after writing the desired folder name after 'cd')
# pwd - It will show Desktop (Its to confirm if its show we are on Desktop or not)
# ls -a -> It will show if repository is available/Which repository is there if available
# mkdir Project -> To create a folder type 'mkdir space the required folder name'
# git init -> By this we can add Git Repository (.git)
# ls -a -> It must show now that .git repository is there

# if we give command (ls) to the newly created folder it will not show anything

# Now to create a file into the folder which is created we have 2 commands:- 
    1. vim f1.py - It will open the file into the vim editor
    2. Now press 'i' button after which we can only write into the vim editor (It will show 'Insert' at bottom after pressing 'i')
    3. Now to exit from the vim editor after typing anything - esc, :wq 
    4. cat f1.py - It will show all contents written in the vim editor f1.py file
    5. To add a single word into the existing file we can use command - echo Hello > f2.py
    6. To add multiple line or sentence use command - vim f1.py
    7. >> - To write scentence in the existing file while using echo command 
    8. git status - It shows the nature of file (Tracked in Green colour & Untracked in Red colour)
    
    ->> Till here all is in the Work Directory
    
Now steps for moving all into the local repository/Staging area:-
        
        1. git add f1.py f2.py/ git add *.py/ git add .
        2. git status (the files colour must come in green colour now)
        3. git commit -m 'Any message'-> 
        4. git status
        
Now steps to move all into the coloud Repository:-
         
        1. git config --global user.name 'sachin-alt'
        2. git config --global user.email 'sachin.sharma14321@gmail.com'
        3. git config --global core.autocrlf true
        4. git remote add origin _and paste the url of rpository from github
        5. git remove -v
        6. git push origin master
    
```
